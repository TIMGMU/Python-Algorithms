### Do NOT use any built-in hash functions or something you download. Create this yourself. 


```python
def Hash (inline, mFactor = 1123):
    """Hash function - takes an input line and returns the HT code. """
    # insert code to convert an input line string into a number that ranges from
    # 0 to the capacity of the hash table HT
    transact_line=''.join(inline)
    hash_num = 0
    for i in transact_line:
        hash_num += ord(i) * mFactor   #Turns string into int
    return hash_num % HTcapacity            #Returns remainder as hash
```

The hash function that I implemented in this assignment utilizes a given python library function ord(), which returns the ASCII unicode code (an integer) of a given character. The function first concatenates all strings in the parameter inline, going through the transaction line from text file "AS6input.txt" and turning every line into a single string for hashing. After the character (or key) is converted into an integer in the for loop, each are then multiplied my mFactor - which represents a constant that provides randomness and flexibility in the hash function. This in turn can affect the distribution of how keys are spread on the hash table and collissions. This step is repeated for every character in the string and essentially follows H1:(key -> integer) which converts a key (or character) into a integer (which also represents the character - vice versa) making it ready for part 2 of hashing, compression.

The compression part of the hash function follows the formula of H2: integers -> [0, N-1] where HTcapacity is N. It's fairly simple in its execution using the modulo of the H1 integer (our hash) and HTcapacity, which is the capacity of our hash table. The purpose of using modulo is to compress the bloated hash values into a more manageable number that lies within a specifc range. Since the first part of hashing requires mapping of the characters into integers, which in most cases wouldd produce a very long integer that is hard to work with. And since we're working with a fixed size slot - it is even more important to use compression to map values into slots. The main purpose is to have an even distribution in accordance to their values allowing for efficient use of memory to optimize operations such as insert, search, and remove.

Collisions occur when inputs have the same hash value which were handled by chaining together records - entries with the same values were stored in linked strings within slots of the hash table. The code basically checks if the slot is none, if it is then insert the record into this empty slot of the hash table and then print. If the slot is not empty, then appened the new record into the existing slot where a chain of strings is 'linked' into the same slot and print "Collision" to notify.
