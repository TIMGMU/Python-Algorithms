## IT309 - using Matplotlib to create a graph

### This notebook is to be used for viewing the various graphs for problem #3 when students aren't able to run the graphs in their own spaces.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.linspace(0, 10, 100)     # generate 100 sample points in range 1 to 10
# Generate lines on the graph
plt.plot(n, 8*n, label = '8n: linear')
plt.plot(n, 2*n**2, label = '2n**2: quadratic')
plt.plot(n, n**3, label = 'n**3: cubic')
plt.plot(n, 2**n, label = '2**n: exponential')
# Labels
plt.title('textrbook: R3.1 - comparison plot of various growth functions')
plt.legend()
# Do it!
plt.show()
```
