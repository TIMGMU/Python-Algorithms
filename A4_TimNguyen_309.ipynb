```python
class SinglyLinkedList:
  """LIFO Stack implementation using a singly linked list for storage."""

 #-------------------------- nested _Node class --------------------------
  class _Node:         # "inner" class
    """Lightweight, nonpublic class for storing a singly linked node."""
    __slots__ = '_element', '_next'           # streamline memory usage

    def __init__(self, element, next):        # initialize node's fields
        self._element = element               # reference to user's element
        self._next = next                     # reference to next node

    def setNext(self, next):
        self._next = next
    
    def getNext(self):
        return self._next
  
    def setElement(self, element):
        self.element = _element
    
    def getElement(self):
        return self._element
        
  #------------------------------- stack methods -------------------------------
  def __init__(self, head = None, size = 0):
    """Create an empty list."""
    self._head = head                       # reference to the head node
    self._size = size                       # number of stack elements

  def getSize(self):
    """Return the number of elements in the list."""
    return self._size

  def isEmpty(self):
    """Return True if the list is empty."""
    return self._size == 0

  def add (self, element):
    """Add element element to the end of the list."""
    node = self._Node(element, None)
    if self.isEmpty():
        self._head = node
    else:
        cursor = self._head
        while cursor.getNext() is not None:
            cursor = cursor.getNext( )
        cursor.setNext(node)
    print('Adding # ', self._size, ' element: ', node.getElement())
    self._size += 1
        
  def remove(self):
    """Remove the element at the head of the list. Raise Empty if the list is empty."""
    if self.isEmpty():
      raise Empty('List is empty')
    self._size -= 1
    out = self._head.getElement()      # save the head element and return to the caller
    self._head = self._head.getNext()  # Reset the head to the second (next) list element
    return out

  def showList(self):
    """Display the contents of the list, from head to tail. Raise Empty exception if the List is empty. """
    if not self.isEmpty():
        cursor = self._head
        while True:
            print(cursor.getElement(), ' reference: ', id(cursor), id(cursor.getNext()))
            if cursor.getNext() == None: break
            cursor = cursor.getNext()
    return
        
  def search(self, key):
    """ Search the list looking for a node element = key. return True if found. """        
    cursor = self._head      # start search at head - set loop pointer cursor to head
    while cursor:
        if key == cursor.getElement():
            return True
        cursor = cursor.getNext()    # point at next node - reset cursor next
    return False

#---------------------------------------------------------------------------------
# IT-309 A4 requires you to complete the code for the two methods that follow
#---------------------------------------------------------------------------------

  def insertAfter (self, i, element):
    """Insert an element after the ith element of the list."""
    node = self._Node(element, None)
  # insert method body below
  # Cases: (1) insert into empty list, (2) insert before the head, (3) insert in the interior
    if i < 0:                #if less than 0, seek pointer at beginning of index...
         i = 0               # to insert before the head
    elif i> self._size:       #if greater than 0, seek pointer towards the end..... 
        i = self._size       #sets i to self.size to be inserted after the tail
    if i == 0:               # insert at beginning since it's a NN
        node.setNext(self._head.getNext())  #insert before the head
        self._head.setNext(node)
    else:
        cursor =self._head    
        for p in range(i-1):       #iterates pointer through node until...
            cursor=cursor.getNext()   #pointer finds next element
        node.setNext(cursor.getNext()) #(3) insert in the interior, sets the NN to desired spot
        cursor.setNext(node)      #points the next pointer to NN
    self._size += 1               #increments size after inserting NN
    

  def insertElt (self, insrtThis, location, insrtHere):
    """Insert element 'insrtThis' before/after node with element =='insrtHere'.
       Parameter 'where' = "B" (before) or "A' (After) 'insrtHere'.        """
    node = self._Node(insrtThis, None)  #New node created
    # insert method body below
    # Hint: search for the list node element == 'insrtHere', then insert accordingly...'  
    cursor = self._head     #pointer set
    prev = None             #previous set

    while cursor:
        if cursor.getElement()== insrtHere:    #if pointer matches insrtHere....
            if location == "B":                #if == B then insert NN before insrtHere
                if prev is None:               #insert before the node with insrtHere
                    node.setNext(self._head)   #insert before the head
                    self._head = node          
                else:                         #else... [prev is NOT none]
                    node.setNext(cursor)      #insert in interior, set new node to point to current node
                    prev.setNext(node)        #set next pointer to previous node  
                self._size += 1               #inccrement after inserting NN
                return       
        elif location == "A":                 #insert after insrtHere pointer
            node.setNext(cursor.getNext())    #set next pointer of NN to point to the node after pointer node
            cursor.setNext(node)              #sets next pointer of current to NN   
            self._size +=1                    #increments after NN insert
            return

        prev= cursor                          #set prev to current pointer
        cursor = cursor.getNext()             #set pointer to get next

def process_transactions(filename, sll):
    with open(filename, "r") as file:
        for line in file:
            node = line.strip().split(',')
            transaction_code = node[0]         #gets first element from list node to represent transaction_code (A, R, IA, IE, SL)

            if transaction_code == "A":         #adds <string value to rear/tail> of the SLL:
                element = node[1]
                sll.add(element)
            elif transaction_code == "R":       #removves the head of SLL
                sll.remove()
            elif transaction_code == "IA":      #creates and inserts a NN with string value as element immediate after sll element # i
                i = int(node[1])                #i = index of which element to be inserted after
                element = node[2]
                sll.insertAfter(i,element)
            elif transaction_code == "IE":   #creates and inserts a NN with element <string1> immediately before or after an existing SLL node whose element is <string2>
                insrtThis = node[1]          #insert before or after another element?
                location= node[2]
                insrtHere= node[3]
                sll.insertElt(insrtThis, location, insrtHere)  #displays all SLL elements from head to tail 
            elif transaction_code == "SL":
                sll.showList()
                

class Empty(Exception):
    """Empty exception class provided to flag that condition. """
    pass
```
